```python
q = QuantumRegister(1)
c = ClassicalRegister(1)
circ = QuantumCircuit(q,c)
circ.x(q[0])
circ.measure(q[0], c[0])
circ.draw(output='mpl')
```

```python
q = QuantumRegister(3)
c = ClassicalRegister(2)
circ = QuantumCircuit(q,c)
circ.x(q[0])
circ.x(q[1])
#circ.cx(q[0],q[2])
circ.ccx(q[0],q[1],q[2])
circ.cx(q[0],q[1])
circ.measure(q[1], c[0])
circ.measure(q[2], c[1])

backend = Aer.get_backend('qasm_simulator')
job = execute(circ, backend)
result = job.result()
#Get results of program
job_stats = result.get_counts()
print(job_stats)
circ.draw(output='mpl')
```

```python
nbit = 2
a = QuantumRegister(nbit)
b = QuantumRegister(nbit)
carIn = QuantumRegister(nbit)
out = QuantumRegister(nbit+1)
cl = ClassicalRegister(nbit+1)
circ = QuantumCircuit(a,b,carIn,out,cl)
for ib in range(nbit):
    circ.cx(a[ib],out[ib])
    circ.cx(b[ib],out[ib])
    circ.ccx(a[ib],b[ib],out[ib+1])
    circ.measure(out[ib], cl[ib])
    circ.measure(out[ib+1], cl[ib+1])

circ.draw(output='mpl')
```
